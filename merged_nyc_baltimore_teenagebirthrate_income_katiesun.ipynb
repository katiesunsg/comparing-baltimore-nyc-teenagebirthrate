# Merging Datasets

```
import pandas as pd
import numpy as np
```

```
# baltimore teenage birth rates by all income
dfbalti_all_teenagebirths = pd.read_csv("https://raw.githubusercontent.com/katiesunsg/comparing-baltimore-nyc-teenagebirthrate/master/balti-allincome-teenagebirthrate.csv")

# baltimore teenage birth rates by high income
dfbalti_high_teenagebirths = pd.read_csv("https://raw.githubusercontent.com/katiesunsg/comparing-baltimore-nyc-teenagebirthrate/master/balti-highincome-teenagebirthrate.csv")

# baltimore teenage birth rates by middle income
dfbalti_middle_teenagebirths = pd.read_csv("https://raw.githubusercontent.com/katiesunsg/comparing-baltimore-nyc-teenagebirthrate/master/balti-middleincome-teenagebirthrate.csv")

# baltimore teenage birth rates by low income
dfbalti_low_teenagebirths = pd.read_csv("https://raw.githubusercontent.com/katiesunsg/comparing-baltimore-nyc-teenagebirthrate/master/balti-lowincome-teenagebirthrate.csv")

# nyc teenage birth rates by all income
dfnyc_all_teenagebirths = pd.read_csv("https://raw.githubusercontent.com/katiesunsg/comparing-baltimore-nyc-teenagebirthrate/master/nyc-allincome-teenagebirthrate.csv")

# nyc teenage birth rates by high income
dfnyc_high_teenagebirths = pd.read_csv("https://raw.githubusercontent.com/katiesunsg/comparing-baltimore-nyc-teenagebirthrate/master/nyc-highincome-teenagebirthrate.csv")

# nyc teenage birth rates by middle income
dfnyc_middle_teenagebirths = pd.read_csv("https://raw.githubusercontent.com/katiesunsg/comparing-baltimore-nyc-teenagebirthrate/master/nyc-middleincome-teenagebirthrate.csv")

# nyc teenage birth rates by low income
dfnyc_low_teenagebirths = pd.read_csv("https://raw.githubusercontent.com/katiesunsg/comparing-baltimore-nyc-teenagebirthrate/master/nyc-lowincome-teenagebirthrate.csv")
```

# Merging Datasets

```
# rename all column 
dfbalti_all_teenagebirths = dfbalti_all_teenagebirths.rename(columns={"Teenage_Birth_Rate_women_only_rP_gF_pall":"Teenage Birth Rate"})
```

```
# preview
dfbalti_all_teenagebirths.head()
```

```
# rename high column 
dfbalti_high_teenagebirths = dfbalti_high_teenagebirths.rename(columns={"Teenage_Birth_Rate_women_only_rP_gF_p75":"Teenage Birth Rate"})
```

```
# preview
dfbalti_high_teenagebirths.head()
```

```
# merge balti all and balti high
df_balti_allhigh = pd.merge(dfbalti_all_teenagebirths, 
						dfbalti_high_teenagebirths, 
						how = "left",
 on = "Teenage Birth Rate")

# preview merge
df_balti_allhigh.head()
```

```
# rename column for middle income
dfbalti_middle_teenagebirths = dfbalti_middle_teenagebirths.rename(columns={"Teenage_Birth_Rate_women_only_rP_gF_p50":"Teenage Birth Rate"})
```

```
# preview 
dfbalti_middle_teenagebirths.head()
```

```
# merge balti all high and middle
df_balti_allhighmiddle = pd.merge(df_balti_allhigh, 
						dfbalti_middle_teenagebirths, 
						how = "left",
 on = "Teenage Birth Rate")
```

```
# preview data
df_balti_allhighmiddle.head()
```

```
# rename low income column 
dfbalti_low_teenagebirths = dfbalti_middle_teenagebirths.rename(columns={"Teenage_Birth_Rate_women_only_rP_gF_p25":"Teenage Birth Rate"})
```

```
# preview
dfbalti_low_teenagebirths.head()
```

```
# merge balti all high middle and low
dfbaltimerged_teenagebirths = pd.merge(df_balti_allhighmiddle, 
						dfbalti_low_teenagebirths, 
						how = "left",
 on = "Teenage Birth Rate")
```

```
# preview merged datasets
dfbaltimerged_teenagebirths.head()
```
